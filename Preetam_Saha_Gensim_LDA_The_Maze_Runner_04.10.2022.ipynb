```python
review_text = dataset.Reviews.apply(gensim.utils.simple_preprocess)
```

```python
len(review_text)
```

```python
review_text
```

```python
review_text[0]
```

```python
model = gensim.models.Word2Vec(
        window = 6,
        min_count=2,
        workers = 4
)
```

```python
model.build_vocab(review_text,progress_per=1000)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.train(review_text,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.save("the_maze_runner_dataset.model")
```

```python
model.wv.most_similar("scary")
```
