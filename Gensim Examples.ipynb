```python
tokens = [list(gensim.utils.tokenize(rev)) for rev in only_reviews1]
tokens
```

```python
dictionaries = gensim.corpora.Dictionary(tokens)
```

```python
dictionaries.filter_extremes(no_below=10)
```

```python
len(list(dictionaries))
```

```python
dictionaries.save("/home/subhasis/Dataset/Web Crawl data/Gensim_Maze_runner/maze_runner.dict") # Use your own path
```

```python
maze_runner_corpora = [dictionaries.doc2bow(t) for t in tokens]
```

```python
gensim.corpora.MmCorpus.serialize("/home/subhasis/Dataset/Web Crawl data/Gensim_Maze_runner/maze_runner.mm",
                                  maze_runner_corpora) # Use your own path
```

```python
tfidf_maze_runner = gensim.models.TfidfModel(maze_runner_corpora)
tfidf_corpus = tfidf_maze_runner[maze_runner_corpora]
```

```python
lsi_model = gensim.models.lsimodel.LsiModel(tfidf_corpus,num_topics= 5, id2word= dictionaries)
```

```python
lsi_model.print_topics(num_topics=5)
```
