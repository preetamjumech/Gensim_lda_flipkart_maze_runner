```python
import pandas as pd
```

```python
dataset = pd.read_csv("The Maze Runner.csv",index_col=0)
dataset.head()
```

```python
only_reviews = dataset.Reviews
```

```python
only_reviews1 = [gensim.utils.any2utf8(rev.lower()) for rev in only_reviews]
only_reviews1
```

```python
only_reviews1 = [gensim.parsing.remove_stopwords(rev) for rev in only_reviews1]
only_reviews1
```

```python
tokens = [list(gensim.utils.tokenize(rev)) for rev in only_reviews1]
tokens
```

```python
len(tokens)
```

```python
model = gensim.models.Word2Vec(window=6,min_count=2,workers=4)
```

```python
model.build_vocab(tokens,progress_per=1000)
```

```python
model.train(tokens,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar("bad")
```

```python
model.wv.similarity("superb","amazing")
```

```python
model.wv.similarity("product","life")
```

```python
dictionaries = gensim.corpora.Dictionary(tokens)
dictionaries
```

```python
len(dictionaries)
```

```python
dictionaries.filter_extremes(no_below=10)
```

```python
len(dictionaries)
```

```python
dictionaries.save("maze_runner_dict.dict")
```

```python
maze_runner_corpora = [dictionaries.doc2bow(t) for t in tokens]
```

```python
maze_runner_corpora
```

```python
len(maze_runner_corpora)
```

```python
tfidf_maze_runner = gensim.models.TfidfModel(maze_runner_corpora)
tfidf_corpus = tfidf_maze_runner[maze_runner_corpora]
```

```python
lsi_model = gensim.models.lsimodel.LsiModel(tfidf_corpus,num_topics=5,id2word=dictionaries)
```

```python
lsi_model.print_topics(num_topics=5)
```
